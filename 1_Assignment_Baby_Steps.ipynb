```python
# Task-3 : BODY MASS INDEX
# BMI 18.5=Under Weight
# BMI 18.5 to 24.9 = Normal Weight
# BMI 25 to 29.9 = Over Weight
# BMI 30 and above = OBESE

# Get BMI input from the user.
BMI=float(input("Enter BMI"))
# Use if-elif-else to categorise
if(BMI<18.5):
    print("Under Weight")
elif(BMI<24.9):
    print("Normal Weight")
elif(BMI<29.9):
    print("Over Weight")
else:
    print("Very Over Weight")
# End of Program
```
