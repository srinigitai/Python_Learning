```python
# Create a function that checks if a given number is odd or even.

# CODE:
def odd_even(): # Defining the function.
    input_number=int(input("Enter a Number: ")) # Getting the input from user by using python input() function. Define the input as integer.
    result=input_number%2 # Checking if the number is divisible by 2 or not and Store the result in a variable.
    if(result==0): # Using if function check if the result is equal to ZERO? If true then print the result.
        print(input_number," is Even Number")
    elif(result!=0): # Using elif function check if the result is not equal to ZERO? If true then print the result.
        print(input_number," is Odd Number")
# End of CODE.
```

```python
# Calling the Function odd_even.
odd_even()
```

```python
# Task: Create a Function that tells Eligibility of marriage for 
#male and female according to their age limit like 21 for male and 18 for female.

# CODE:
def marriage_eligible(): # Define Function
    gender=input("Your Gender: ") # getting the Gender input and store in a variable 'gender'
    age=int(input("Your age: ")) # Getting the age input and store in a variable 'age'

    if(gender=="male"): # Using if function, Check if the gender is 'male'?
        if(age>20): # if the above condition is true, using if function check if the age is greater than 20?
            print("ELIGIBLE") # print if the condition is true.
        else: # if the condition is not matching then print the below message.
            print("NOT ELIGIBLE")

    if(gender=="female"): # using if function check if the gender is 'female'?
        if(age>17): # using if function check the age is above 17?
            print("ELIGIBLE") # print if above condition is true.
        else: # if the above condition is not matching then print as below.
            print("NOT ELIGIBLE")
# END OF CODE.
```

```python
marriage_eligible() # Calling the function.
```

```python
# Task: Calculate the percentage of your 10th mark

def tenth_percentage(): # Defining the function.
    # Getting the subject marks as inputs in five variables.
    sub1=98
    sub2=87
    sub3=95
    sub4=95
    sub5=93
    
    # adding all the 5 subject marks and storing it to a variable.
    total_marks=sub1+sub2+sub3+sub4+sub5
    # Calculating the percentage of the marks.
    percent=total_marks/5
    
    # Printing the marks, total and percentage.
    print("Subject 1= ",sub1)
    print("Subject 2= ",sub2)
    print("Subject 3= ",sub3)
    print("Subject 4= ",sub4)
    print("Subject 5= ",sub5)
    print("Total = ",total_marks)
    print("Percentage= ",percent)
# End of code
```

```python
tenth_percentage() # Calling the function.
```
