```python
# Task: Print 'CORRECT' if i==10

# CODE:
# Get input from the user using a variable 'i'. Compare if the input is equal to 10. If the input is equal to 10 then print 'CORRECT'

i=int(input("Enter a Number: i= "))
if(i==10):
    print("CORRECT")
# End of CODE
```

```python
# Task: Check the passward, using if and else
# Enter the password: HOPE@123
# Your Password is Correct.

# CODE:
# Declare a variable and define a password. 
# Declare a variable and get the input from the user.
# Compare the user entered password with the defined password.
# If both are matching then print the result. If not matching print the result.

pwd="HOPE@123"
user_pwd=input("Enter the Password: ")
if(user_pwd==pwd):
    print("Your Password is Correct")
else:
    print("Incorrect Password")
# End of CODE
```

```python
# Task: Categorize the people by their age like child,adult,citizing and Senior Citizen.

# CODE:
# Get the user input using a variable "age"
age=int(input("Enter the Age: "))
# Using if - elif statement check the input "age" value. Catagorize based on the input.31
if(age<18):
    print("Age ",age," is a Child")
elif(age<35):
    print("Age ",age," is a Adult")
elif(age<60):
    print("Age ",age," is a Citizen")
else:
    print("Age ",age," is a Senior Citizen")
# End of CODE
```

```python
# Task: Find wheather the given number is Positive or negative?

# CODE:
# Get the input number from the user using a variable.
# Check if the entered value is greater than ZERO or Less than ZERO?
# If the entered value is greater then ZERO then print as POSITIVE and if it is less than ZERO Print it as NEGATIVE.

num=int(input("Enter a Number: "))
if(num>=0):
    print("The Number is POSITIVE")
elif(num<0):
    print("The Number is NEGATIVE")
# End of CODE
```

```python
# Check wheather the given number is divisible by 5.

# CODE:
# Get the input number from the using a variable.
# if the Modulo of the number is = zero then it is divisible by 5. 
num_5=int(input("Enter the Number: ")) # num_5 is a variable declared to get the input from the user.
if(num_5%5==0): # % is used to check if the reminder is zero.
    print(num_5," is Divisible by 5")
else:
    print(num_5, " is not Divisible by 5")
```
