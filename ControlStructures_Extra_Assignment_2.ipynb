```python
# Task:Print 0 to 20 using range.
# Use range() function to do this. range() function works for a range of numbers.
# range() function has three input parameters. 1. Start Range, 2. End Range and 3. Increment steps which is optional. If increment not given then default is 1.
# Syntax: for temp variable in range(start range,end range,[increment steps])

# CODE:
# seq is temporary variable.
for seq in range(0,20):
    print(seq)
# End of CODE.
```

```python
# print range 10 to 20

# CODE:
# seq1 is temporary variable.
for seq1 in range(10,20):
# Normally, the print statement gives the results one by one in a new line while printing.
# To print it in the same line we can use an end statement which will end printing in next line and print it in the same line.
    print(seq1,end=" ")
# End of CODE
```

```python
# Task: Print number of items in the list by using "len"
# [10,20,14,55,43,87,76]
# Output: Number of items in the list2: 7
# Syntax: len(variable). This len function counts the number of items in the variable.

# CODE:
list2=[10,20,14,55,43,87,76]
num_items=len(list2)
print("Number of items in the list2: ",num_items)
# End of Code
```

```python
# Task: Print the word "Artificial Intellignece" Horizontally in one line and print every letter in this word one by one in a new line.
# For horizontal printing We can use a print statement.
# For vertical printing we can use a list and then print that list using for loop.

# CODE:
alpha_list=["A","r","t","i","f","i","c","i","a","l"," ","I","n","t","e","l","l","i","g","e","n","c","e"]
print("Artificial Intelligence")
for temp_list in alpha_list:
    print(temp_list)
# End of Code
```

```python
# Task: Print Odd Number in the List. (20,10,16,19,25,1,276,188)

# CODE:
num_list=[20,10,16,19,25,1,276,188]
print("(",20,10,16,19,25,1,276,188,")")
for div in num_list:
    if (div%2!=0):
        print(div," is ODD Number")
# End of Code
```

```python
# Task: Print Even Number in the List. (20,10,16,19,25,1,276,188)

# CODE:
num_list=[20,10,16,19,25,1,276,188]
print("(",20,10,16,19,25,1,276,188,")")
for div in num_list:
    if (div%2==0):
        print(div," is Even Number")
# End of Code
```
